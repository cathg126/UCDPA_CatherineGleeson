
Analysing data - merge DataFrames


```python
sub_exdf1 = exdf[exdf['Activity, Exercise or Sport (1 hour)'].str.contains('mile')]
sub_exdf1
```

```python
sub_exdf2 = exdf[exdf['Activity, Exercise or Sport (1 hour)'].str.contains('Walk')]
sub_exdf2
```

```python
sub_exdf3 = sub_exdf2[sub_exdf2['Activity, Exercise or Sport (1 hour)'].str.contains('mph')]
sub_exdf3
```

```python
dfg = pd.concat([sub_exdf1,sub_exdf3], axis=0)
dfg
```

# Visualisations & Insights

1. Demonstrating calories burned by walking and running speed for a 205 lb person

```python
ax2 = dfg.plot.barh(x='Activity, Exercise or Sport (1 hour)', y='205 lb')
```

Insight: This graph visually demonstrates the increase in calories burned while walking and running as speed increases. It is also interesting to see how visualising the data highlights the significant amount of additional calories burned while running as against those burned while walking.

2. Plotting the calories burned by a 205 lb runner as their speed increases

```python
sub_exdf1.plot.line(y='205 lb')
```

Insight: It is clear from the graph that the calories burned by the runner increased at a steady rate as their running speed increased.

2. Plotting the comparative calories burned by runners of varying weights as their speed increases

```python
plt.subplots(figsize=(10,10))

plt.plot(sub_exdf1['205 lb'],sub_exdf1['Activity, Exercise or Sport (1 hour)'])
plt.plot(sub_exdf1['180 lb'],sub_exdf1['Activity, Exercise or Sport (1 hour)'])
plt.plot(sub_exdf1['155 lb'],sub_exdf1['Activity, Exercise or Sport (1 hour)'])
plt.plot(sub_exdf1['130 lb'],sub_exdf1['Activity, Exercise or Sport (1 hour)'])

plt.xlabel('Calories burned')
plt.ylabel('Running speed (mph)')
plt.title('Increase in calories burned by running speed and weight')
```

Insight: By graphing the available data, it is clear that, regardless of the starting weight of the runner, their calories burned will increase consistently as their running speed increases.
